```python
dataset = tablib.Dataset().load(open('dataset.csv').read())
```

```python
print(dataset)
```

```python
def minMax(dataset):
    
    minimum = min(map(int,dataset['age']))
    maximum = max(map(int,dataset['age']))
    return minimum,maximum
```

```python
def countPeople(dataset,lower,upper): 
    count = 0
    for i in map(int,dataset['age']):
        if i<= upper and i >= lower:
            count += 1
    return count
    
```

```python
def countOver(dataset,bound):
    count = 0
    for i in map(int,dataset['age']):
        if i>= bound :
            count += 1
    return count
```

```python
def countLastName(dataset,lastname):
    count = 0
    for i in dataset['last_name']:
        if i == lastname :
            count += 1
    return count
```

```python
import numpy as np
def avgIncome(dataset):
    return np.average(list(map(int,dataset['2017_gross_income_usd'])))
```

```python
def getDistinctOccupation(dataset):
    return list(set(dataset['employment_status']))
```

```python
def avgIncomeWithAge(dataset,lower,upper):
    listOfIncome = list()
    for i in dataset.dict:
        if int(i['age'])<= upper and int(i['age']) >= lower:
            listOfIncome.append(int(i['2017_gross_income_usd']))
    return np.average(listOfIncome)
    
```

```python
def avgGender(dataset,gender):
    listOfIncome = list()
    for i in dataset.dict:
        if i['gender'] == gender:
            listOfIncome.append(int(i['2017_gross_income_usd']))
    return np.average(listOfIncome)

def genderGap(dataset):
    return avgGender(dataset,'M') - avgGender(dataset,'F')
```

```python
def findRichestZip(dataset):
    listOfIncome = dict()
    for i in dataset.dict:
        if int(i['2017_gross_income_usd']) not in listOfIncome.keys():
            listOfIncome[int(i['2017_gross_income_usd'])] = [i['zip']]
        else:
            listOfIncome[int(i['2017_gross_income_usd'])].append(i['zip'])
    maxIncome = max(list(listOfIncome.keys()))
    return listOfIncome[maxIncome]
```

```python
def findHighestLowest(dataset):
    education = list(set(dataset['education']))
    employment = list(set(dataset['employment_status']))
    dictionary = dict()
    for i in education:
        for j in employment:
            key = (i,j)
            temp = []
            for k in dataset.dict:
                if i == k['education'] and j == k['employment_status']:
                    temp.append(int(k['2017_gross_income_usd']))
            average = np.average(temp)
            if average not in dictionary.keys():
                dictionary[average] = [key]
            else:
                dictionary[average].append(key)
    low = min(list(dictionary.keys()))
    high = max(list(dictionary.keys()))
    return dictionary[low],dictionary[high]            
```
